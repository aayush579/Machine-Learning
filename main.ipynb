```python
fig = px.histogram(medical_df, x="age",
                   marginal='box',
                   nbins=47,
                   title='Age Distribution')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.histogram(medical_df, x="bmi",
                   marginal='box',
                   color_discrete_sequence=['red'],
                   title='BMI Distribution')
fig.update_layout(bargap=0.1)
fig.show()
```
